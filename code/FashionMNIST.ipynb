```python
from Ethan import *
from MyCNN import *
import matplotlib.pyplot as plt
```

```python
batch_size = 256
train_dataloader, test_dataloader = FashionMNIST_DataLoader(batch_size)

text_labels = datasets.FashionMNIST.classes

fig, axes = plt.subplots(2, 5, figsize=(12, 5))
for i, ax in enumerate(axes.flat):
    j = torch.randint(len(train_dataloader.dataset), size=(1,)).item()
    ax.imshow(train_dataloader.dataset[j][0][0])
    ax.set_title(text_labels[train_dataloader.dataset[j][1]])
    ax.set_xticks([])
    ax.set_yticks([])
```

```python
net = LeNet_BatchNorm()
model = MyNetwork(net)
```

```python
epochs = 35
lr = 1e-3

for t in range(epochs):
    print(f"Epoch {t+1}\n----------------")
    train(train_dataloader, model, lr)
    test(test_dataloader, model)
print("Done!")

# torch.save(model.state_dict(), "../model/FashionMNIST.pth")
# model.load_state_dict("../model/FashionMNIST_best.pth")
```
